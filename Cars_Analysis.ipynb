**Q1.) Find all the NULL values in the dataset. If there is any NULL Values in any column,then fill it with mean of that column.**

```python
df.isnull().sum()
```

From above output we come to know that there are two null values in the "Cylinders" column we have to fill that 2 null values with the mean of that column.

```python
df['Cylinders'].fillna(df['Cylinders'].mean(),inplace=True)
```

The NULL values has been filled with the mean of the column and is permanently mmade changes to the dataset.

```python
df.isnull().sum()
```

Now, We dont see any NULL value in the columns.
